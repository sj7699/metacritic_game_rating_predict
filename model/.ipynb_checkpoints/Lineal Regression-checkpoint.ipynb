장르별 개수조사(여러 장르를 포함한경우 계수를 주기위해서)
개수가 낮을수록 장르의 특수성을 고려하여 가중치를 더 줄예정

```python
dic_genre={}
for idx,row in df_genres.iterrows():
    nowg=row['genres']
    nowg=nowg.split('|')
    nowg.pop()
    for x in nowg:
        if(x in dic_genre):
            dic_genre[x]+=1
        else:
            dic_genre[x]=0
```
